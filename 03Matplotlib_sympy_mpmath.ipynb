## Tasas de convergencia

Si tenemos una secuencia $a_n$ que converge a $a$ cuando $n \to \infty$:

Calculamos el *error* $\epsilon_n  := a_n - a$. Si $a_n \to a$, entonces $\epsilon_n \to 0$. ¿Qué tan rápido? O sea, ¿cómo varía $\epsilon_n$ con respecto a $n$ cuando $n \to \infty$?

Por ejemplo, $a_n = 1 - 1 / n \to 1$.  $\epsilon_n = -1/n = n^{-1}$

Gráficamente:

```python
n = np.arange(1, 100)
a_n = 1 - 1./ n
```

```python
n
```

```python
a_n
```

```python
plt.plot(n, a_n, 'o-')
```

```python
epsilon_n = a_n - 1
plt.plot(n, epsilon_n)
```

```python
plt.loglog(n, abs(epsilon_n), 'o-')

plt.loglog(n, n**(-3.))
```
